# Task 1
В стране Х все, кто рождается в феврале, становятся программистами. При этом особенно высоко ценятся программисты, рожденные 29 февраля. Известно, что такая дата – редкость. Многие мошенники указывают ложную дату рождения, чтобы получить больше уважения и почестей, поэтому правительство страны Х решило дополнительно проверять данные, которые предоставляют программисты при устройстве на любую работу. 

Известно, что мошенники почему-то не знают, что 29 февраля бывает только в високосный год, благодаря чему их можно вычислить. Помогите правительству написать программу, которая среди людей, указавших дату рождения в феврале, выделяет мошенников, обычных программистов и особо ценных программистов. 

**Входные данные:** два целых числа, каждое вводится с клавиатуры в отдельной строке. Первое число – день рождения (целое число от 1 до 29), второе число – год рождения, целое положительное число.

**Вывод программы:** одна из трех строк: «мошенник» (если указано 29 февраля, а год при этом не високосный), «обычный программист» (если рожден не 29 февраля), «особо ценный программист» (если рожден 29 февраля високосного года).

**Примеры**

Входные данные:
10
1998
Вывод программы:
обычный программист

Входные данные:
29
2000
Вывод программы:
особо ценный программист


Входные данные:
29
1900
Вывод программы:
мошенник

```python
day = int(input())
year = int(input())

if day != 29:
    print("обычный программист")
elif ((year % 4) == 0 and (year % 100) != 0) or (year % 400 == 0):
    print("особо ценный программист")
else:
    print("мошенник")
```

* Конь ходит на две клетки по вертикали в любом направлении и на одну клетку по горизонтали, или наоборот.


**Входные данные:** четыре целых числа, от 1 до 8 каждое. Каждое число вводится в отдельной строке. Первая пара чисел – номер строки и столбца первой клетки, вторая пара – номер строки и столбца второй клетки.

Вывод программы:

Входные данные:

Ладья не может совершить такой ход
Слон не может совершить такой ход

```python
x_start, y_start = int(input()), int(input())
x_finish, y_finish = int(input()), int(input())

distance_x = abs(x_finish - x_start)
distance_y = abs(y_finish - y_start)
    
is_achievable_by_rook = distance_x == 0 or distance_y == 0

is_achievable_by_bishop = distance_x == distance_y

is_achievable_by_knight = (
    (distance_x == 2 and distance_y == 1)
    or (distance_x == 1 and distance_y == 2)
)

if is_achievable_by_rook:
    print("Ладья может совершить такой ход")
else:
    print("Ладья не может совершить такой ход")
    
if is_achievable_by_bishop:
    print("Слон может совершить такой ход")
else:
    print("Слон не может совершить такой ход")
    
if is_achievable_by_knight:
    print("Конь может совершить такой ход")    
else:
    print("Конь не может совершить такой ход")
```

# Task 3

Напишите программу, определяющую тип треугольника (прямоугольный, остроугольный или тупоугольный) по его сторонам. Учтите, что треугольник с заданными сторонами может не существовать вовсе.

**Входные данные:** три целых неотрицательных числа, по одному в каждой строке

**Вывод программы:** одно из четырех сообщений – “Треугольник с такими сторонами не существует”, “Прямоугольный треугольник”, “Остроугольный треугольник”, “Тупоугольный треугольник” в зависимости от типа треугольника.

**Примеры**

Входные данные:
3
5
4
Вывод программы:
Прямоугольный треугольник

Входные данные:
100
1
2
Вывод программы:
Треугольник с такими сторонами не существует 

```python
a, b, c = int(input()), int(input()), int(input())

if a >= b and a >= c:
    a, c = c, a
elif b >= a and b >= c:
    b, c = c, b    
    
if a >= b + c or b >= a + c or c >= a + b:
    print("Треугольник с такими сторонами не существует")
else:
    a_square, b_square, c_square = a ** 2, b ** 2, c ** 2
    
    if c_square > a_square + b_square:
        print("Тупоугольный треугольник")
    elif c_square < a_square + b_square: 
        print("Остроугольный треугольник")
    else:
        print("Прямоугольный треугольник")
```

# Task 4

На координатной прямой расположены два отрезка. В случае, если они пересекаются более, чем в одной точке, вычислите длину пересечения отрезков. Если отрезки пересекаются в одной точке, выведите координату точки пересечения. Если же отрезки не пересекаются вовсе, выведите сообщение об этом. Используйте формат вывода из примеров.

**Входные данные:** четыре целых числа, вводятся по одному в каждой строке. Числа задают две пары — координаты начала и конца первого отрезка в первой паре, во второй аналогично. Гарантируется, что первое число в паре строго меньше второго.

**Вывод программы:** одно из трех сообщений:

* Отрезки не пересекаются
* Отрезки пересекаются в точке x (x – точка пересечения)
* Длина пересечения отрезков y (y – длина пересечения отрезков)

**Примеры**

Входные данные:
-1
0
2
5
Вывод программы:
Отрезки не пересекаются
 

Входные данные:
1
5
3
1000
Вывод программы:
Длина пересечения отрезков 2


Входные данные:
-123
4
4
9
Вывод программы:
Отрезки пересекаются в точке 4

```python
left_border_1, right_border_1 = int(input()), int(input()) 
left_border_2, right_border_2 = int(input()), int(input())

if left_border_1 > left_border_2:
    left_border_1, left_border_2 = left_border_2, left_border_1
    right_border_1, right_border_2 = right_border_2, right_border_1

if left_border_1 == left_border_2:
    if right_border_1 <= right_border_2:
        print(f"Длина пересечения отрезков {right_border_1 - left_border_1}")
    else: 
        print(f"Длина пересечения отрезков {right_border_2 - left_border_2}")
else:
    if right_border_1 > left_border_2:
        if right_border_2 < right_border_1:
            print(f"Длина пересечения отрезков {right_border_2 - left_border_2}")
        else:
            print(f"Длина пересечения отрезков {right_border_1 - left_border_2}")
    elif right_border_1 == left_border_2:
        print(f"Отрезки пересекаются в точке {right_border_1}")
    else:
        print("Отрезки не пересекаются")
```
