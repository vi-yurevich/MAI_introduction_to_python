# Task 1

Найдите второй максимум и второй минимум в последовательности чисел. Гарантируется, что все числа в последовательности различны, и что в последовательности не менее двух чисел. При решении задачи запрещено использовать сортировку списка – обрабатывайте числа потоково, не сохраняя их в список.

**Примечание:** пусть numbers – отсортированный по возрастанию список чисел. Тогда второй минимум – элемент с индексом 1, второй максимум – элемент с индексом len(numbers) - 2. 

**Входные данные:** целые положительные числа, вводятся по одному в каждой строке. В качестве символа окончания последовательности вводится число 0.

**Вывод программы:** два числа, по одному в каждой строке. Первое – второй минимум среди элементов введенной последовательности, второе – второй максимум.

**Примеры**

Входные данные:
1
9
2
5
0
Вывод программы:
2
5

Входные данные:
10
2
0
Вывод программы:
10
2

````python
first_number = int(input())
second_number = int(input())
minimum, maximum = min(first_number, second_number), max(first_number, second_number)
pred_min, pred_max = maximum, minimum

while True:
    next_number = int(input())
    
    if next_number == 0:
        break
    
    if next_number < pred_min:
        pred_min = next_number
        
        if pred_min < minimum:
            minimum, pred_min = pred_min, minimum
    
    if next_number > pred_max:
        pred_max = next_number
        
        if pred_max > maximum:
            maximum, pred_max = pred_max, maximum
        
print(pred_min)
print(pred_max)
````

# Task 2

Дана последовательность чисел. Определите, сколько в этой последовательности элементов, которые строго больше двух своих соседей, и выведите количество таких элементов. Крайние элементы последовательности (первое и последнее число в ней) никогда не учитываются, ведь у них недостаточно соседей.

**Входные данные:** целые числа, вводятся в одной строке через пробел.

**Вывод программы:** одно число, количество подходящих под условия элементов последовательности.

**Пример** 

Входные данные:
-1 5 2 1
Вывод программы:
1

**Пояснение к примеру:** в этой последовательности только число 5 строго больше двух своих соседей – чисел -1 и 2.

````python
numbers = [int(number) for number in input().split(' ')]
numbers_len = len(numbers)

count = 0
for (i, j, k) in zip(range(0,numbers_len - 2), range(1, numbers_len - 1), range(2, numbers_len)):
    if numbers[i] < numbers[j] and numbers[j] > numbers[k]:
        count += 1

print(count)
````

# Task 3

В Python предусмотрен перевод чисел между некоторыми системами счисления при помощи встроенных функций, например, можно перевести десятичное число в двоичную форму записи при помощи функции bin(). Однако не предусмотрено встроенной функции, которая позволяла бы перевести число из десятичной системы счисления в другую с произвольным основанием. Реализуйте перевод числа, записанного в десятичной системе счисления, в систему счисления с основанием base, где 2<=base<=16. Для записи цифр в системах счисления с основанием больше 10 используйте прописные латинские буквы: A, B, C, D, E, F.

**Входные данные:** два натуральных числа, вводятся по одному в каждой строке. Первое число — число, которое необходимо перевести из десятичной системы счисления. Второе — base, основание системы счисления, в которую осуществляется перевод.

**Вывод программы:** строка, представление введенного числа в системе счисления с основанием base.

**Примеры**

Входные данные:
10
2
Вывод программы:
1010

Входные данные:
100
15
Вывод программы:
6A

````python
number = int(input())
base = int(input())

quotient = number
result = ""

while quotient >= base:
    remainder = quotient % base
    quotient = quotient // base
    if base > 10:
        if remainder == 10:
            remainder = "A"
        if remainder == 11:
            remainder = "B"
        if remainder == 12:
            remainder = "C"
        if remainder == 13:
            remainder = "D"
        if remainder == 14:
            remainder = "E"
        if remainder == 15:
            remainder = "F"
    result += str(remainder)
else:
    result += str(quotient)
    result = result[::-1]

print(result)
````

# Task 4

НОК (наименьшее общее кратное) и НОД (наибольший общий делитель) играют важную роль в теории чисел и построении некоторых вычислительных алгоритмов. Реализуйте вычисление НОД и НОК для двух натуральных чисел.

**Входные данные:** два натуральных числа, вводятся по одному в каждой строке.
**Вывод программы:** cтроки в формате “НОД введенных чисел: x”, “НОК введенных чисел: y”, где x, y — вычисленные НОД и НОК соответственно.

**Примеры**

Входные данные:
1234
5678
Вывод программы:
НОД введенных чисел: 2
НОК введенных чисел: 3503326

Входные данные:
12
35
Вывод программы:
НОД введенных чисел: 1
НОК введенных чисел: 420

````python
x = int(input())
y = int(input())

a, b = x, y

while b > 0:
    a, b = b, a % b
    
gcd = a
lcm = x * (y // gcd)

print(f"НОД введенных чисел {gcd}")
print(f"НОK введенных чисел {lcm}")
````

# Task 5

Напишите программу, которая выводит фрагмент таблицы умножения для чисел из диапазона [x, y] на числа из диапазона [w, z], где x, y, w, z – натуральные числа, не превосходящие 10. Обратите внимание на форматирование вывода, ориентируйтесь на пример.

**Входные данные:** в первой строке вводятся числа x, y, во второй строке ввода – w, z. Числа в строке разделены пробелом. Первое из пары чисел гарантированно меньше второго.

**Вывод программы:** фрагмент таблицы умножения. При этом первая строка и первый столбец – заголовочные, в них содержатся сами числа из диапазонов. Числа в рамках каждой строки находятся в ячейках длины 7 с выравниванием по правому краю (дополняются слева пробелами до длины в 7 символов).

**Пример**

Входные данные:
2 5
3 6
Вывод программы:
```text
             3      4      5      6
      2      6      8     10     12
      3      9     12     15     18
      4     12     16     20     24
      5     15     20     25     30
```

````python
input_numbers = [int(i) for i in input().split(' ') + input().split(' ')]
x, y, w, z = input_numbers[0], input_numbers[1], input_numbers[2], input_numbers[3]

result = [[""]]
for i in range(x, y + 1):
    result.append([str(i)])
    
for j in range(w, z + 1):
    result[0].append(str(j))

for (row, i) in zip(result[1:], range(x, y + 1)):
    for j in range(w, z + 1):
        row.append(str(i * j))

for row in result:
    for i in row:
        print(i.rjust(7),end='')
    print()
````

# Task 6

Для регистрации в системе пользователю необходимо придумать достаточно сложный пароль. Напишите программу, которая проверяет, достаточно ли сложный пароль придумал пользователь. Если пользователь не выполнил какие-то условия, накладываемые на допустимый (достаточно сложный для регистрации) пароль, необходимо сообщить пользователю об этом.

Требования к паролю:
-длина не менее 10 символов
-пароль содержит хотя бы одну букву в верхнем регистре
-пароль содержит хотя бы одну букву в нижнем регистре
-пароль содержит хотя бы одну цифру
-пароль содержит хотя бы один буквенный символ
-пароль не содержит пробелов
-пароль не содержит повторяющихся символов

**Входные данные:** одна строка.

**Вывод программы:** одна из двух строк – “Вы ввели допустимый пароль” или “Вы ввели недопустимый пароль”.

**Примеры**

Входные данные:
hello
Вывод программы:
Вы ввели недопустимый пароль
 
Входные данные:
Ab12345678
Вывод программы:
Вы ввели допустимый пароль

Входные данные:
PYTHON123454321
Вывод программы:
Вы ввели недопустимый пароль

````python
password = input()

if (
    len(password) < 10
        or password.isnumeric()
        or password.isascii()
        or password.isupper()
        or password.islower()
        or " " in password):
    isValid = False
else:
    isValid = True
    
if not isValid:
    for i, char in enumerate(password):
        isValid = not char in password[i:]

if isValid:
    print("Вы ввели допустимый пароль")
else:
    print("Вы ввели недопустимый пароль")
````
