# Task 1

В качестве входных данных программе подается шестизначное десятичное число. Выведите числа:
1. Составленное из цифр, стоящих на четных позициях в данном числе, записанных в обратном порядке.
2. Составленное из цифр, стоящих на нечетных позициях в данном числе.
3. Произведение двух вышеуказанных чисел.

Все результирующие числа должны быть выведены на одной строке с символом “&” в качестве разделителя.

**Входные данные:** одно целое положительное число, вводится с клавиатуры.
**Вывод программы:** три числа в одной строке, разделенных символом “&”

**Пример**

Входные данные: 
123456
Вывод программы: 
642&135&86670

**Примечание:** в решении задачи запрещено использовать строки, циклы, списки, условный оператор if, операторы сравнения и т.д. Используйте только целые числа и операции с ними.

```python
number = int(input())

zero = number % 10
first = number % 100 // 10
second = number % 1_000 // 100
third = number % 10_000 // 1_000
fourth = number % 100_000 // 10_000
fifth = number % 1_000_000 // 100_000

first_result_number = zero * 100 + second * 10 + fourth 
second_result_number = fifth * 100 + third * 10 + first
third_result_number = first_result_number * second_result_number

print(zero, second, fourth, '&', sep='', end='')
print(fifth, third, first, '&', sep='', end='')
print(third_result_number)
```

# Task 4

На вход подается шестизначное положительное целое число. Гарантируется, что число действительно шестизначное, то есть в нем не ведущих нулей.  Определите, является ли число симметричным. Если число симметричное, требуется вывести 1, иначе 0.

**Входные данные:** одно целое положительное шестизначное число
**Вывод программы:** одно целое число (0 или 1 в зависимости от симметричности введенного шестизначного)

**Примеры**

Входные данные: 
123456
Вывод программы: 
0

Входные данные: 
245542
Вывод программы: 
1

**Примечание:** в решении задачи запрещено использовать строки, циклы, списки, условный оператор if, операторы сравнения и т.д. Используйте только целые числа и операции с ними.

```python
number = int(input())

zero = number % 10
first = number % 100 // 10
second = number % 1_000 // 100
third = number % 10_000 // 1_000
fourth = number % 100_000 // 10_000
fifth = number % 1_000_000 // 100_000

compare_digits = (zero ^ fifth) ^ (first ^ fourth) ^ (second ^ third)
not_is_palindrome = min(compare_digits, 1)
is_palindrome = (not_is_palindrome + 1) % 2

print(is_palindrome)
```

# Task 5

Пользователем последовательно вводятся положительные вещественные числа x, y, z, каждое вводится с клавиатуры на отдельной строке. Вычислите для них значение функции.
$x^2\times\left(\sqrt{y}+\sqrt{z}\right)\times\frac{1}{\sqrt{x^6+\sqrt{y}}}-\sqrt{3x}+z^{11}$

Выведите результат с тремя знаками после запятой с округлением. (используйте форматирование строк)

**Входные данные:** три вещественных положительных числа, по одному в каждой строке

**Вывод программы:** одно вещественное число

**Пример**

Входные данные: 
1.1
0.2
0.1
Вывод программы: 
-1.196

```python
x = float(input())
y = float(input())
z = float(input())

result = x ** 2 * (y ** 0.5 + z ** 0.5) * (((x ** 6 + y ** 0.5) ** 0.5) ** (-1)) - ((3 * x) ** 0.5) + z ** 11

print(f"{result:.3f}")
```
