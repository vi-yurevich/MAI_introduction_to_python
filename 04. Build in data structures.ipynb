# Task 1

Пользователь вводит текст с клавиатуры. Предоставьте пользователю статистику по введенному тексту: количество различных слов в нём, количество употреблений каждого из слов в этом тексте.
**Примечание:** В этой задаче словом можете считать любую последовательность символов, отделенную пробелом от соседних.

**Входные данные:** одна строка с текстом, которую пользователь вводит с клавиатуры. Строка не содержит символов препинания.

**Вывод программы:** в первой строке вывода – количество различных слов в нем в формате «В вашем тексте n различных слов». В последующих строках – информация о частотности каждого слова в формате «Слово x встретилось в вашем тексте n раз».

**Пример**

Входные данные:
текст с повторами слов часто звучит хуже чем текст без повторов слов
Вывод программы:
В вашем тексте 10 различных слов
Слово текст встретилось в вашем тексте 2 раз
Слово с встретилось в вашем тексте 1 раз
Слово повторами встретилось в вашем тексте 1 раз
Слово слов встретилось в вашем тексте 2 раз
Слово часто встретилось в вашем тексте 1 раз
Слово звучит встретилось в вашем тексте 1 раз
Слово хуже встретилось в вашем тексте 1 раз
Слово чем встретилось в вашем тексте 1 раз
Слово без встретилось в вашем тексте 1 раз
Слово повторов встретилось в вашем тексте 1 раз

```python
string = input()
words_counts = {}

for word in string.split(' '):
    if word in words_counts:
        words_counts[word] += 1
    else:
        words_counts[word] = 1

print(f"В вашем тексте {len(words_counts)} различных слов")
for word, count in words_counts.items():
    print(f"Слово {word} встретилось в вашем тексте {count} раз")
```

# Task 2

Часто на практике возникает задача определить уникальность текста. В этом задании вам необходимо подсказать пользователю, насколько его текст уникален по сравнению с текстом-образцом. Уникальность будем определять, как процентное отношение количества различных слов, которые встречаются в тексте пользователя, но не встречаются в тексте-образце, к общему количеству различных слов в тексте пользователя.

**Примечание:** В этой задаче словом можете считать любую последовательность символов, отделенную пробелом от соседних.
 

**Входные данные:** две строки с текстами, каждая вводится с клавиатуры на отдельной строке. В строках не содержится знаков препинания. Первая строка – текст-образец. Вторая строка – текст, который надо проверить на уникальность.

**Вывод программы:** первой строке вывода – информация о количестве общих слов в тексте-образце и тексте пользователя. Во второй строке – уникальность, задана как в условии задачи, измеряется в процентах. Уникальность требуется округлить до целого значения. 

 

**Пример**

Входные данные:
это текст образец текст состоит из некоторого количества слов
текст с повторами слов часто звучит хуже чем текст без повторов слов
Вывод программы:
2
80%

**Примечание:** здесь в тексте-образце и в тексте пользователя два общих слова – «текст» и «слов». Общее количество различных слов во втором предложении – 10, поэтому уникальность – 80%.

```python
sample_words = {word for word in input().split(' ')}
user_words = {word for word in input().split(' ')}

number_of_identical_words = abs(len(sample_words) - len(user_words))

percentage_of_uniqueness = round((len(sample_words.union(user_words)) - len(sample_words)) / len(user_words) * 100)

print(number_of_identical_words)
print(percentage_of_uniqueness, '%', sep='')
```

# Task 3

В стране Y люди не знают о существовании арабских цифр и все числа записывают прописью. Помогите реализовать калькулятор, которым эти люди смогли бы пользоваться. Людей в стране Y обрадует даже простейший калькулятор, который позволит им складывать два числа, каждое из которых находится в диапазоне от 0 до 10 включительно и вычитать из одного числа в этом диапазоне другое. Калькулятор должен предоставлять результат в понятном людям формате – итоговое число тоже должно быть записано прописью.


**Входные данные:** одна строка формата «число1 плюс число2» или «число1 минус число2», вводится с клавиатуры.

**Вывод программы:** одна строка с суммой, записанной прописью.

**Примеры**

Входные данные:
пять плюс ноль
Вывод программы:
пять

Входные данные:
семь плюс девять
Вывод программы:
шестнадцать

Входные данные:
семь минус девять
Вывод программы:
минус два

Входные данные:
восемь минус один
Вывод программы:
семь

```python
numbers_words_pairs = {
    0 : "ноль",
    1 : "один",
    2 : "два",
    3 : "три",
    4 : "четыре",
    5 : "пять",
    6 : "шесть",
    7 : "семь",
    8 : "восемь",
    9 : "девять",
    10 : "десять",
    11 : "одиннадцать",
    12 : "двенадцать",
    13 : "тринадцать",
    14 : "четырнадцать",
    15 : "пятнадцать",
    16 : "шестнадцать",
    17 : "семнадцать",
    18 : "восемнадцать",
    19 : "девятнадцать",
    20 : "двадцать",
}

words_numbers_pairs = {}
for number, word in numbers_words_pairs.items():
    words_numbers_pairs[word] = number
    
expression = input().split(' ')
if expression[1] == 'плюс':
    result_number = words_numbers_pairs[expression[0]] + words_numbers_pairs[expression[2]]
if expression[1] == 'минус':
    result_number = words_numbers_pairs[expression[0]] - words_numbers_pairs[expression[2]]
    
if result_number >= 0:    
    print(numbers_words_pairs[result_number])
else:
    print('минус', numbers_words_pairs[-result_number], sep=' ')
```

# Task 4

Проверьте, является ли матрица, введенная пользователем, одной из "особенных" матриц.

Необходимо проверить следующие свойства матрицы:

 * матрица является верхнетреугольной (квадратная матрица, все элементы, стоящие ниже главной диагонали -- нули)
 * матрица является диагональной (квадратная матрица, все элементы которой, стоящие не на главной диагонали -- нули, на главной диагонали могут быть какие угодно числа)
 * матрица является единичной (диагональная матрица, на главной диагонали которой все элементы -- единицы).

**Входные данные:** первой строке вводятся два натуральных числа через пробел, m и n, где m -- количество строк в матрице, n -- количество столбцов. m > 1, n > 1

Далее вводятся m строк, в каждой из которых перечислены целые числа -- элементы очередной строки матрицы

**Вывод программы:** одно из четырех сообщений:
  * Матрица верхнетреугольная
  * Матрица диагональная
  * Матрица единичная
  * Матрица не является особенной

**Примечание 1:** При решении задачи запрещено пользоваться numpy и прочими библиотеками. При необходимости используйте встроенные структуры данных Python

**Примечание 2:** Необходимо вывести только одно сообщение из перечисленных, самую "строгую" характеристику матрицы. Диагональная матрица -- частный случай верхнетреугольной, считаем её диагональной. Аналогично единичная матрица -- частный случай диагональной.

```python
m, n = [int(number) for number in input().split(' ')]
matrix = []

for i in range(m):
    matrix.append([int(number) for number in input().split(' ')])
    
is_upper_triangular = False
is_diagonal = False
is_identity = False

if n == m:
    unique_elements_below_diag = list({
        matrix[i][j] 
        for i in range(n)
        for j in range(i)
    })
    
    if (len(unique_elements_below_diag) == 1 
            and unique_elements_below_diag[0] == 0):
        is_upper_triangular = True
        unique_elements_above_diag = list({
                matrix[i][j]
                for i in range(n)
                for j in range(i + 1, n)
        })
    
    if (is_upper_triangular 
            and len(unique_elements_above_diag) == 1 
            and unique_elements_above_diag[0] == 0):
        is_diagonal = True    
        unique_elements_on_diag = list({ matrix[i][i] for i in range(n) })
        
    if (is_diagonal 
            and len(unique_elements_on_diag) == 1 
            and unique_elements_on_diag[0] == 1):
        is_identity = True
        
if is_identity:
    print("Матрица единичная")
elif is_diagonal:
    print("Матрица диагональная")
elif is_upper_triangular:
    print("Матрица верхнетреугольная")
else:
    print("Матрица не является особенной")     
```
